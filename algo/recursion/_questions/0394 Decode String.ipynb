```python
def decodeString(s):
    res, num = '', 0
    def helper(s, i):
        while i < len(s):
            if s[i].isdigit():
                num = num * 10 + int(s[i])
            elif s[i].isalpha():
                res += s[i]
            elif s[i] == '[': #recurse
                res += helper(s) * num
                num = 0
            elif s[i] == ']':
                return res
        return res
    return helper(s, 0)
```
