### 144 Binary Tree Preorder Traversal

```python
def preorderTraversal(root):
    if not root:
        return []
    
    result = []
    
    def preorder(node):
        result.append(node.val)   #root first
        if node.left:
            preorder(node.left)
        if node.right:
            preorder(node.right)

    preorder(root)
    return result
```

iterative: start with root in the stack. At every iteration, pop the current node out of stack and push its child nodes into the stack (first right and then left). In this implementation, nodes are pushed into output list in the following order:  Top->Bottom and then Left->Right => producing preorder traversal. (similar to BFS but with stack so right child pushed first and then left child since stack is LIFO)

```python
def preorderTraversal2(root):
    if not root:
        return []
    
    stack, result = [root], []
    
    while stack:
        node = stack.pop()
        result.append(node.val)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    return result
```

```python
def preorderTraversal3(root):
    if not root:
        return []
    
    stack = [(root,None)]
    result = []
    
    while len(stack) != 0:
        (node, zone) = stack[-1]
        
        if zone is None:
            stack[-1] = (node, "arrival")
            result.append(node.val)
            if node.left:
                stack.append((node.left, None))
        
        elif zone == "arrival":
            stack[-1] = (node, "interim")
            if node.right:
                stack.append((node.right, None))
        
        elif zone == "interim":
            stack[-1] = (node, "departure")
            stack.pop()
        
    return result
```

```python
preorderTraversal(root)
```

### 94 Binary Tree Inorder Traversal

```python
def inorderTraversal(root):
    if not root: 
        return []
    
    result = []

    def inorder(node):
        if node.left:
            inorder(node.left)
        result.append(node.val)   #root in middle
        if node.right:
            inorder(node.right)

    inorder(root)
    return result
```

```python
inorderTraversal(root)
```

### 145 Binary Tree Postorder Traversal

```python
def postorderTraversal(root):
    if not root: 
        return []
    
    result = []

    def postorder(node):
        if node.left:
            postorder(node.left)
        if node.right:
            postorder(node.right)
        result.append(node.val)   #root last

    postorder(root)
    return result
```

```python
postorderTraversal(root)
```
